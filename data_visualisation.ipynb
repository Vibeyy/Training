```python
#handling for null values
import numpy as np
dataset.head()
dataset.isnull().sum()
columns = dataset.columns
for col in columns:
    if np.nan == dataset[col][0]:
        i = 1
        while dataset[col][i] == np.nan:
            i += 1
        dataset[col][0] = dataset[col][i]
    if np.nan == dataset[col][len(dataset)-1]:
        i=len(dataset-2)
        while dataset[col][i] == np.nan:
            i-=1
        dataset[col][len(dataset)-1] = dataset[col][i]

dataset.bfill(inplace=True)
dataset.isnull().sum()


```

```python
#remove outliers

dataset_copy = dataset.copy()
for column in dataset_copy.select_dtypes(include = 'float64'):
    q1 = dataset_copy[column].quantile(0.25)
    q3 = dataset_copy[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    dataset_copy = dataset_copy[(dataset_copy[column] > lower_bound) & (dataset_copy[column] < upper_bound)]
    
sns.boxplot(dataset_copy['bill_length_mm'])

```

```python
import pandas as pd
from functools import reduce
def find_avg(col):
    if dataset[col].dytpe in ['float64','int64']:
        return dataset_copy[col].mean()
    else:
        return "Invalid data type"

def find_sum(col,penguin_name = None):
    if dataset[col].dtype in ['float64','int64']:
        if penguin_name != None:
            return dataset_copy[col][dataset_copy['species'] == penguin_name].sum()
        else:
            return dataset_copy[col].sum()
    else:
        return "Invalid data type"
    
def find_count(val):
    if len(dataset[dataset['species'] == val]) == 0:
        return "Value not found"
    else:
        return len(dataset[dataset['species'] == val])
    

def get_species_count(species):
    filtered = list(filter(lambda x: x == species, dataset['species']))
    return len(filtered)


```
